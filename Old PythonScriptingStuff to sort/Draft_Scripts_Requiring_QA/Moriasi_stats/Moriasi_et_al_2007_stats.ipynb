```python
#
gbr.init('D:/Test/scen_RC7',
         obs_directory='D:/a_sc_p2r_b_rc7/Data/WQ_Observed/Daily_TS/',
         timeperiods='D:/a_sc_p2r_b_rc7/Data/WQ_Observed/Daily_TS/Timeseries_class.csv')
gbr.available()
results = gbr.Results('cc')
site = '120006b'
obs = results.load_observed_ts(site)
pred = results.get_node_ts(site,obs.columns)
```

```python
pred ['TN'] = pred['N_Particulate'] + pred['N_DIN'] + pred['N_DON']
pred ['TP'] = pred['P_Particulate'] + pred['P_FRP'] + pred['P_DOP']
pred [0:3]
```

```python
site = '12006bTNTP'
obs = results.load_observed_ts(site)
```

```python
df1 = gbr.stats.nse(obs, pred)
df1
```

```python
df2 = gbr.stats.PBIAS(obs, pred)
```

```python
df3 = gbr.stats.ioad(obs, pred)
```

```python
df4 = gbr.stats.rae(obs, pred)
```

```python
df5 = gbr.stats.rsr(obs, pred)
```

```python
frames_daily = [df1, df2, df3, df4, df5]
result_daily = pd.concat(frames_daily, keys=['NSE','PBIAS', 'ioad', 'rae', 'rsr'])
result_daily = pd.DataFrame(result_daily)
```

```python
result_daily.columns = [site]
result_daily
```

```python
obs_month = obs.groupby([lambda x: x.year, lambda x: x.month]).sum()
```

```python
obs_pred = pred.groupby([lambda x: x.year, lambda x: x.month]).sum()
```

```python
df6 = gbr.stats.nse(obs_month, obs_pred)
```

```python
df7 = gbr.stats.PBIAS(obs_month, obs_pred)
```

```python
df8 = gbr.stats.ioad(obs_month, obs_pred)
```

```python
df9 = gbr.stats.rae(obs_month, obs_pred)
```

```python
df10 = gbr.stats.rsr(obs_month, obs_pred)
```

```python
frames_monthly = [df6, df7, df8, df9, df10]
result_monthly = pd.concat(frames_monthly, keys=['NSE','PBIAS', 'ioad', 'rae', 'rsr'])
result_monthly = pd.DataFrame(result_monthly)
```

```python
result_monthly.columns = [site]
result_monthly
```

```python
result = pd.concat([result_daily, result_monthly], axis=1, join='inner')
result
```
