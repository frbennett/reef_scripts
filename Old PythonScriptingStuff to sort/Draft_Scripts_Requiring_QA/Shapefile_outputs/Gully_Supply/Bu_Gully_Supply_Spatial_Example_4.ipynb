```python
# Some steps required until these Python modules are properly installed...
import sys
sys.path.append('../../Modules')
sys.path.append('../../../../veneer-py')
# Get the Source scripting components (veneer) and GBR specific stuff
import gbr
import fiona
%pylab inline
import pandas as pd
import geopandas as gpd
#
```

```python
df = pd.read_csv("d:/a_sc_p2r_b_rc7/Model_runs/BD_1/Runs/a_scen_reader/RC7/cc/RawResults.csv")
# Filter raw results gully load data
df = df[(df["ModelElementType"]=='Catchment') & (df["BudgetElement"]=='Gully') & (df["Process"]=='Supply')]
df [0:3]
```

```python
df = pd.pivot_table(df,index=["ModelElement","FU"],values=["Total_Load_in_Kg"],columns=["Constituent"])
df [0:3]
```

```python
df = df.xs('Total_Load_in_Kg', axis=1, drop_level=True)
df [0:3]
```

```python
df = df/1000000
df
```

```python
b_intsFu = gpd.read_file('D:/a_sc_p2r_b_rc7/Model_runs/BD_1/Configuration/bu_intSCFU.shp')
b_intsFu.rename(columns={'reef_luse':'FU','IntSCFU':'DROP','Column_0':'ModelElement'},inplace=True)
b_intsFu.set_index(['ModelElement','FU'],inplace=True)
b_intsFu [0:3]
```

```python
spatialLoads = b_intsFu.join(df)
spatialLoads.rename(columns={'N_Particulate':'PN','P_Particulate':'PP','Sediment - Coarse':'SC','Sediment - Fine':'FS'},inplace=True)
spatialLoads [0:5]
```

```python
spatialLoads.geometry = gpd.GeoSeries(spatialLoads.geometry)
```

```python
spatialLoads.to_file('D:/a_sc_p2r_b_rc7/Model_runs/BD_1/Runs/a_scen_reader/RC7/CC/CL_hillsope_fs.shp')
```
