```python
grouped_rawResults = raw.reset_index().groupby(['Process', 'BudgetElement', 'Constituent']).sum()
```

```python
#massBalanceKg = grouped_rawResults.reset_index().pivot(['Process','BudgetElement'], 'Constituent', 'Total_Load_in_Kg')
massBalanceKg = pd.pivot_table(grouped_rawResults.reset_index(),index = ['Process','BudgetElement'], columns = 'Constituent', values = 'Total_Load_in_Kg')
```

```python
#massBalanceKgAnnual = massBalanceKg/results.runDetails.yearsOfRecording
massBalanceKgAnnual = massBalanceKg/28.4
```

```python
massBalanceAnnual = massBalanceKgAnnual.copy()
conversions = {'Sediment - Fine':1e-6,'Flow':1e-6,'N_DIN':1e-3,'N_DON':1e-3,
'N_Particulate':1e-3,'P_DOP':1e-3,'P_FRP':1e-3,'P_Particulate':1e-3}
for col in massBalanceAnnual.columns:
    if col in conversions:
        massBalanceAnnual[col] *= conversions[col]
massBalanceAnnual
```
