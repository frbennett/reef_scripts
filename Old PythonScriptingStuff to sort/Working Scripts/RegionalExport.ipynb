```python
# How its done...

# 1) Load the regional summary table
allRegional = results.get('RegionalSummaryTable')
allRegional
```

```python
# 2) Filter just the Export values
regionalExportData = allRegional[allRegional.MassBalanceElement=='Export']
regionalExportData
```

```python
# 3) Pivot the table to be Constituent x Region
# 3-a) Convert index (Constituent) back to a normal column
regionalExportData = regionalExportData.reset_index()
regionalExportData
```

```python
# 3-b) Pivot  
regionalExportKg = regionalExportData.pivot('Constituent','SummaryRegion','Total_Load_in_Kg')
# Pivot is called with (index,columns,values), 
# so ('Constituent','SummaryRegion', 'Total_Load_in_Kg') means...
# Make Contituent the index (the rows),
# Make SummaryRegion the columns, and
# Use Total_Load_in_Kg as the values
regionalExportKg
```

```python
# 4) Convert from kg to T/y
# 4-a) Kg -> T
regionalExport_T = regionalExportKg / 1000.0
regionalExport_T
```

```python
# 4-b) Calculate Simulation Length
simulationLength = results.runDetails.end - results.runDetails.startRecording
simulationLength
```

```python
# 4-c) Simulation length in years
years = simulationLength.days / 365.25
years
```

```python
# 4-d) Convert T->T/y
regionalExport_T_y = regionalExport_T/years
regionalExport_T_y
```

```python
# 5) Export to ___
regionalExport_T_y.to_csv('D:/beckers_regional_export.csv')
regionalExport_T_y.to_excel('D:/beckers_regional_export.xlsx')
```
