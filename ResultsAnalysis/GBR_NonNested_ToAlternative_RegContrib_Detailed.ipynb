```python
### This will be the routine used to take a non-nested GBR regional contributor table and re-assign regions
### Must be using Non-Nested raw data


import pandas as pd
import numpy as np

selectedConstituents = ['Sediment - Fine', 'N_DIN', 'Ametryn', 'Atrazine', 'Diuron', 'Hexazinone', 'Tebuthiuron']

stdGenString = 'LoadToStream (kg)'
stdExpString = 'LoadToRegExport (kg)'
numDaysString = 'Num_Days'
repRegString = 'Rep_Region'
fuString = 'FU'
aream2String = 'AreaM2'
processString = 'Process'
RSDRstring = 'RSDR'
constitString = 'Constituent'

origContribStringForCatch = 'ModelElement'
desiredStringForCatch = 'Catchment'

#rawcontrib = pd.read_csv('P:\projects\RC8_ResultsSets\BU\Regional_Contributor\BU_RC2016_Baseline_RegContributorDataGrid.csv')
#outputFile = 'P:\projects\RC8_ResultsSets\BU\Regional_Contributor\BU_RC2016_Baseline_RegContributorDataGrid_annualised_FilteredWeber_AUSGOV.csv'

rawcontrib = pd.read_csv('P:\projects\RC8_ResultsSets\FI\Regional_Contributor\FI_RC2016_Baseline_RegContributorDataGrid.csv')
outputFile = 'P:\projects\RC8_ResultsSets\FI\Regional_Contributor\FI_RC2016_Baseline_RegContributorDataGrid_annualised_FilteredWeber_AUSGOV.csv'


tempStringForAnnGen = 'AnnLoadToStreamKG'
tempStringForAnnExp = 'AnnLoadToExportKG'


rawcontrib[tempStringForAnnGen] = rawcontrib[stdGenString].div(rawcontrib[numDaysString]).mul(365.25)
rawcontrib[tempStringForAnnExp] = rawcontrib[stdExpString].div(rawcontrib[numDaysString]).mul(365.25)
#rawcontrib['AreaHA'] = rawcontrib['AreaM2'].div(10000)

### Could put alternatives in here to load a different regionalisation from CSV
#altRegs = pd.read_csv('P:\projects\RC7_Results\BU_Exported ReportingRegions.csv')
altRegs = pd.read_csv('P:\projects\RC7_Results\FI_ReportingRegions_NoCurtisIsland.csv')
### Identify the column names we need
catchColName = 'Catchmt'
regColName = 'AUSGMCAS'
```

```python
### add alternative regions to raw contributor data

rawcontrib = pd.merge(rawcontrib, altRegs, how='left', left_on=['ModelElement'], right_on=[catchColName]).reset_index()
rawcontrib.head(5)
```

```python
filtered = rawcontrib.loc[rawcontrib[constitString].isin(selectedConstituents)]

#Need to make the new dataframe a standalone object, not a 'slice' of a copy?????
#Otherwise can't drop columns
filtered = filtered.copy()


filtered.head(5)
```

```python
filtered.drop([stdGenString, stdExpString, repRegString], axis=1, inplace=True)
#filtered.drop([stdGenString], axis=1)
#filtered.drop([stdExpString], axis=1)
#filtered.drop([repRegString], axis=1)

filtered.rename(index=str, columns={regColName: repRegString, tempStringForAnnGen: stdGenString, tempStringForAnnExp: stdExpString, origContribStringForCatch: desiredStringForCatch}, inplace=True)

filtered.head(5)



##Remove original data cols, then rename our temp ones. The ',1' tells DROP we're using columns, not rows
#filtered.drop(stdGenString, 1)
#filtered.drop(stdExpString, 1)
#filtered.drop(repRegString, 1)

#filtered.rename(index=str, columns={regColName: repRegString, tempStringForAnnGen: stdGenString, tempStringForAnnExp: stdExpString, origContribStringForCatch: desiredStringForCatch})

#cols = filtered.columns.tolist()
#cols
```

```python
myCols = [repRegString, desiredStringForCatch, constitString, fuString, processString, stdGenString, stdExpString, RSDRstring, aream2String]

filtered.to_csv(outputFile, columns=myCols, index=False)

print("Done")
```
