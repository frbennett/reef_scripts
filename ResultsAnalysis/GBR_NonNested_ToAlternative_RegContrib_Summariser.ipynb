```python
### This will be the routine used to take a non-nested GBR regional contributor table and produce the summaries
### that we typically supply
### will eventually try to declare as much stuff up front as possible

import pandas as pd
import numpy as np

selectedConstituents = ['Sediment - Fine', 'N_Particulate']
#rawcontrib = pd.read_csv('P:\projects\RC7_Results\FZ_Baseline_RegContributorDataGrid_Rob.csv')
rawcontrib = pd.read_csv('P:\projects\RC8_ResultsSets\BU\RobsRuns\BU_RC8_BL_RegContributorDataGrid_AusGovSource_P2RReg.csv')

rawcontrib['AnnLoadToStreamKG'] = rawcontrib['LoadToStream (kg)'].div(rawcontrib['Num_Days']).mul(365.25)
rawcontrib['AnnLoadToExportKG'] = rawcontrib['LoadToRegExport (kg)'].div(rawcontrib['Num_Days']).mul(365.25)
rawcontrib['AreaHA'] = rawcontrib['AreaM2'].div(10000)

### Could put alternatives in here to load a different regoinalisation from CSV
altRegs = pd.read_csv('P:\projects\RC7_Results\BU_Exported ReportingRegions.csv')

```

```python
rawcontrib.head(5)
```

```python

altRegs.head(15)
```

```python
### Identify the column names we need
catchColName = 'Catchmt'
regColName = 'AUSGMCAS'
```

```python
### Can keep reporting region in the area table, as this is our Non-Nested notebook
#catchFuArea = pd.DataFrame(rawcontrib.groupby(['Rep_Region','ModelElement','FU']).agg({'AreaHA':'first'})).reset_index()
catchFuArea = pd.merge(pd.DataFrame(rawcontrib.groupby(['Rep_Region','ModelElement','FU']).agg({'AreaHA':'first'})).reset_index(), altRegs, how='left', left_on=['ModelElement'], right_on = [catchColName])
```

```python
catchFuArea.head(15)
```

```python
regFuArea = pd.DataFrame(catchFuArea.groupby([regColName,'FU']).agg({'AreaHA':'sum'})).reset_index()
regFuArea.head(25)
```

```python
### add alternative regions to raw contributor data

rawcontrib = pd.merge(rawcontrib, altRegs, how='left', left_on=['ModelElement'], right_on=[catchColName]).reset_index()
rawcontrib.head(5)
```

```python
#regLuseSummary = pd.DataFrame(rawcontrib.groupby(['Rep_Region','Constituent','FU','Process']).agg({'AnnLoadToStreamKG':'sum','AnnLoadToExportKG':'sum'})).reset_index()
regLuseSummary = pd.DataFrame(rawcontrib[rawcontrib['Constituent'].isin(selectedConstituents)].groupby([regColName,'Constituent','FU','Process']).agg({'AnnLoadToStreamKG':'sum','AnnLoadToExportKG':'sum'})).reset_index()
regLuseSummary.head(15)
```

```python
### Join/merge with Regional Areas
regLuseSumPlusArea = pd.merge(regLuseSummary, regFuArea, how='left', left_on=[regColName,'FU'], right_on = [regColName,'FU'])
regLuseSumPlusArea.head(15)
```

```python
regLuseSumPlusArea.to_csv('D:/test_out_Burd_AusGov.csv', index=False)
```
