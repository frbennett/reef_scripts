```python
# Filter CSV By Contents
import pandas as pd
#import pandasql as psql
import glob
import os


reqCols = ['Date', 'Flow m3_per_s', 'Sediment - Fine kg', 'Sediment - Coarse kg', 'N_Particulate kg', 'N_DIN kg', 'N_DON kg', 'P_Particulate kg', 'P_DOP kg', 'P_FRP kg']

basePathSource = r'\\athenasmb\pdreef\Source_Modelling\ereefs\eReefs_Extension_November2019'
basePathDest = r'\\athenasmb\pdreef\Source_Modelling\ereefs\eReefs_Extension_November2019\ConstitFiltered'

regExtDict = {
                'BM':['Baffle_Aggregated_Outflow.csv','Burnett_Aggregated_Outflow.csv',
                      'Burrum_Aggregated_Outflow.csv','Kolan_Aggregated_Outflow.csv','Mary_Aggregated_Outflow.csv'],
               'BU':['Black_Aggregated_Outflow.csv','Burdekin_Aggregated_Outflow.csv',
                     'Don_Aggregated_Outflow.csv','Haughton_Aggregated_Outflow.csv','Ross_Aggregated_Outflow.csv'],
               'CY':['Endeavour River_Aggregated_Outflow.csv','Jacky Jacky Creek_Aggregated_Outflow.csv',
                     'Jeannie River_Aggregated_Outflow.csv','Lockhart River_Aggregated_Outflow.csv',
                     'Normanby River_Aggregated_Outflow.csv','Olive Pascoe_Aggregated_Outflow.csv',
                     'Stewart River_Aggregated_Outflow.csv'],
               'FI':['Boyne_Aggregated_Outflow.csv','Calliope_Aggregated_Outflow.csv',
                     'Shoalwater_Aggregated_Outflow.csv','Styx_Aggregated_Outflow.csv','Water Park_Aggregated_Outflow.csv'],
               'MW':['OConnel_Aggregated_Outflow.csv','Pioneer_Aggregated_Outflow.csv',
                     'PlaneC_Aggregated_Outflow.csv','Prossie_Aggregated_Outflow.csv'],
               'WT':['Barron_Aggregated_Outflow.csv','Daintree_Aggregated_Outflow.csv',
                     'Herbert_Aggregated_Outflow.csv','Johnstone_Aggregated_Outflow.csv',
                     'Mossman_Aggregated_Outflow.csv','Mulgrave-Russell_Aggregated_Outflow.csv',
                     'Murray_Aggregated_Outflow.csv','Tully_Aggregated_Outflow.csv']
              }
```

```python
def doFiltering(inputFile, outputFileDir, outputFileName, theCols):
    
    
    df = pd.read_csv(inputFile)
    
    #filtered = df.loc[df[filterColName].isin(reqVals)]
    if not os.path.exists(outputFileDir):
        os.makedirs(outputFileDir, exist_ok=True)
        print("Made dir: " + outputFileDir)
    
    outputFile = os.path.join(outputFileDir, outputFileName)
    df.to_csv(outputFile, index=False, columns=theCols)

    print("Done " + inputFile)
```

```python
#regBit = '\\RC9_ScenarioBaseline_AggregatedRegionalTS\\BM\\'
topDir = '\\RC9_ScenarioBaseline_AggregatedRegionalTS\\'

for regCode in regExtDict.keys():
    
    regBit = topDir + regCode + '\\'
    
    for csvFile in regExtDict[regCode]:
        inFile = basePathSource + regBit + csvFile
        outFileDir = basePathDest + regBit
        doFiltering(inFile, outFileDir, csvFile, reqCols)

print('All done')
##thisFile = 'Baffle_Aggregated_Outflow.csv'
#inFile = basePathSource + regBit + thisFile
#outFileDir = basePathDest + regBit
#doFiltering(inFile, outFileDir, thisFile, reqCols)
```
