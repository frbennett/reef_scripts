```python
"""Basic Sankey diagrams for modelled fine sediment budgets in the six GBR regions.
"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey
```

The following scripts are modifications of Example 2 in the "Source Code" link under:
http://matplotlib.org/examples/api/sankey_demo_basics.html. It demonstrates:
customised by: Banti Fentie 
to generate fine sediment budgets for each of the six
GBR regions.

Using this code, it is possible to:
- Set one path longer than the others
- Place a label in the middle of the diagram
- Use the scale argument to normalize the flows
- Implicitly pass keyword arguments to PathPatch()
- Change the angle of the arrow heads
- Change the offset between the tips of the paths and their labels
- Format the numbers in the path labels and the associated unit
- Change the appearance of the patch and the labels after the figure is created

The code in each cell below is for a single region.
The name of the region is in the title line with two letter abreviation (e.g. BM = Burnett Mary).
Whilst it has been attempted to generate the budget diagrams between regions as consistently as possible,
there could be differences in how labels appear depending on the proportions of budget eleements included
in the budget.

```python
# This cell has the script to generate TSS budget for the Burnett Mary region
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, xticks=[], yticks=[],
                     title="Flow Diagram of Mackay Whitsunday sediment budget (RC2016)")
sankey = Sankey(ax=ax, scale=0.0003, offset=0.1, head_angle=45,
                format='%.0f', unit=' kt/y')
H_supply=overAll_summary.iloc[10,0]+overAll_summary.iloc[12,0]
G_supply=overAll_summary.iloc[9,0]
B_supply=overAll_summary.iloc[11,0]
CR_supply=overAll_summary.iloc[8,0]
EX_loss=-1*(overAll_summary.iloc[0,0]+
      overAll_summary.iloc[2,0]+
      overAll_summary.iloc[6,0]+
      overAll_summary.iloc[7,0])
FPD_loss=-1*overAll_summary.iloc[1,0]
RD_loss =-1*overAll_summary.iloc[4,0]
SD_loss =-1*overAll_summary.iloc[5,0]
CR_export =-1*(exp_transpose.iloc[0,4])
G_export =-1*(exp_transpose.iloc[2,4])
H_export=-1*(exp_transpose.iloc[5,4]+exp_transpose.iloc[12,4])
B_export =-1*(exp_transpose.iloc[11,4])
sankey.add(flows=[G_supply, H_supply, B_supply, EX_loss, FPD_loss, RD_loss, G_export, H_export, B_export],
           labels=['Gully (G)', 'Hillslope (H)', 'Bank (B)', 'Extraction and\nother losses', '\n\nFloodplain\ndeposition', 'Reservoir\ndeposition', 
                   'Export (G)', 'Export (H)', 'Export (B)'],
           orientations=[1, 1, 0, -1, -1, -1, 1, 1, 0],
           pathlengths=[0.2, 0.25, 0.25, 0.1, 0.2, 0.3,
                        0.15, 0.25, 0.15],
           patchlabel="MW fine sediment budget",
           alpha=0.4, lw=1.0)  # Arguments to matplotlib.patches.PathPatch()
diagrams = sankey.finish()
plt.rc('font', size=14)
diagrams[0].patch.set_facecolor('#37c959')
diagrams[0].text.set_fontsize(14)
diagrams[0].text.set_fontweight('bold')
figManager = plt.get_current_fig_manager()
figManager.window.showMaximized()
plt.text(-1.3,0.2,'TOTAL SUPPLY\n      657 kt/y', fontsize = 20, color = 'black', rotation=90)
plt.text(1.3,0.4,'TOTAL EXPORT\n       626 kt/y', fontsize = 20, color = 'black', rotation=90)
plt.text(0.8,-0.7,'TOTAL STORAGE\n      30 kt/y', fontsize = 20, color = 'black', rotation=0)
plt.show()
plt.savefig('MW_kty.jpg')
```
