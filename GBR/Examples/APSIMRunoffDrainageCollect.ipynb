```python
# Filter CSV By Contents
import pandas as pd
#import pandasql as psql
import glob
import os

```

```python
searchpath = r"E:\ReefSugar_RC10\BM\SOILNUT\baseline"
#searchString = "Runoff"
searchString = "Drainage"
outPath = r"E:\ReefSugar_RC10\BM"
```

```python
counter = 0

combinedDF = pd.DataFrame()

for fullFileName in glob.glob(searchpath + '/*' + searchString + '*'):
    
    fName = os.path.basename(fullFileName)
    nameBits = fName.split('$')
    theCatch = nameBits[0]
    
    thisDF = pd.read_csv(fullFileName)
    thisDF = thisDF.rename(columns={'Date':'Date', searchString+'$mm':theCatch})
    
    if counter == 0:
        #Create dataFrame
        combinedDF = thisDF
    
    else:
        #Add to dataframe
        combinedDF = pd.merge(combinedDF, thisDF, how='left', left_on=['Date'], right_on=['Date'])
    
    counter += 1

print("All data read in to data frame")
```

```python
combinedDF.reset_index()
combinedDF.head(20)
```

```python
outputFile = os.path.join(outPath,searchString+".csv")
combinedDF.to_csv(outputFile, index=False)
print("Done")
```
